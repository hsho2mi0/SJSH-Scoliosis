```python
from PIL import Image
```

```python
route=".\\Image\\"
img_name="img.png"
label_name="label.png"
```

```python
save_train_img_route="./data/train/image"
save_train_label_route="./data/train/label"
save_test_img_route="./data/test/image"
save_test_label_route="./data/test/label"
```

```python
for n in range(1, 181, 1):
    image=Image.open(route+"a%d_json/"%n+img_name)
    image.save(save_train_img_route+"/%d.png"%n)
    print("%d.png complete"%n)
```

```python
for n in range(1, 181, 1):
    image=Image.open(route+"a%d_json\\"%n+label_name)
    image.save(save_train_label_route+"\\%d.png"%n)
    print("%d.png complete"%n)
```
